# Appply some advance functions

```python
s1 = pd.DataFrame(np.random.randint(80,100, (10,5)), columns=["s1","s2","s3","s4","s5"])
s2 = pd.DataFrame(np.random.randint(70,79, (10,5)), columns=["s1","s2","s3","s4","s5"])
s3 = pd.DataFrame(np.random.randint(60,69, (10,5)), columns=["s1","s2","s3","s4","s5"])
s4 = pd.DataFrame(np.random.randint(50,59, (10,5)), columns=["s1","s2","s3","s4","s5"])
s5 = pd.DataFrame(np.random.randint(40,49, (10,5)), columns=["s1","s2","s3","s4","s5"])
s6 = pd.DataFrame(np.random.randint(33,39, (10,5)), columns=["s1","s2","s3","s4","s5"])
s7 = pd.DataFrame(np.random.randint(0,32, (10,5)), columns=["s1","s2","s3","s4","s5"])

df = pd.concat([s1,s2,s3,s4,s5,s6,s7])
df.head()
```

```python
df['obtained'] = df.s1 + df.s2 + df.s3 + df.s4 + df.s5;
df
```

```python
df[["s1", "s2", "s3", "s4", "s5"]].sum() # this will sum whole from top to bottom
```

```python
df['obtained'] = df.loc[:,'s1':'s5'].sum(axis=1) # this will sum left to right 

df
```

```python
def obtained_marks(sub1, sub2, sub3, sub4, sub5):
    return sub1 + sub2 + sub3 + sub4+ sub5

df['obtained'] = df[['s1','s2','s3','s4','s5']].apply(lambda x:obtained_marks(x[0],x[1],x[2],x[3],x[4]), axis=1)
df
```

```python
def calculate(sub1, sub2, sub3, sub4, sub5):
    obt = sub1 + sub2 + sub3 + sub4+ sub5
    per = obt / 500 * 100
    
    grade = ""
    
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "F"
    
    return obt, per, grade

df[['obtained',"percentage","Grade"]] = df[['s1','s2','s3','s4','s5']].apply(lambda x:calculate(x[0],x[1],x[2],x[3],x[4]), axis=1, result_type='expand')
df
```
