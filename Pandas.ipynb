### isin function

```python
df2 = titanicDf[titanicDf.Age.isin([24, 22])]
print(len(df2))
display(df2.head(5))
```

```python
df2 = titanicDf[titanicDf.Name.isin(['Braund, Mr. Owen Harris', 'Sirayanian, Mr. Orsen'])]
print(len(df2))
display(df2.head(5))
```

```python
air_quality = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv", index_col=0, parse_dates=True)

air_quality.head()
```

```python
air_quality.plot()
```

```python
air_quality['station_paris'].plot()
plt.show()
```

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
plt.show()
```
